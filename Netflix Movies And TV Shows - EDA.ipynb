# Analysis

Droping Duplicates 

```python
df.duplicated().sum()
df.drop_duplicates()
df.head(2)
```

```python
df.isna().sum()
```

Droping unnecessary column

```python
df.drop(['description'],axis=1,inplace=True)
df.head()
```

Handling Missing values

```python
df['director'].fillna('unvaliable',inplace=True)
df['cast'].fillna('unvaliable',inplace=True)
df['country']=df['country'].fillna(df['country'].mode()[0])
df['date_added']=df['date_added'].ffill()
df['rating']=df['rating'].fillna(df['rating'].mode()[0])
df.head()
```

Changing Data Type : Int to Date

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(),
                                  format='%B %d, %Y', errors='coerce')
```

```python
df['date_added']
```

```python
df['rating'].unique()
```

```python
valid_ratings = ['PG-13', 'TV-MA', 'PG', 'TV-14', 'TV-PG', 'TV-Y', 'TV-Y7', 'R',
       'TV-G', 'G', 'NC-17', 'NR',
       'TV-Y7-FV', 'UR']

df['rating'] = df['rating'].where(df['rating'].isin(valid_ratings))
```

```python
df
```

```python
df['rating'].unique()
```

```python
df['type'].value_counts()
```

Renaming the columns

```python
df.rename(columns = {'listed_in' : 'categories'}, inplace=True)
df.head()
```

```python
df['duration_min'] = df['duration'].apply(
    lambda x: int(str(x).split()[0]) if isinstance(x, str) and 'min' in x else 0
)
```

```python
df['duration_season'] = df['duration'].apply(
    lambda x: int(str(x).split()[0]) if isinstance(x, str) and 'Seasons' in x else 0
)
```

# Visualization

```python
# Plot the distribution of first 10 countries 
plt.figure(figsize=(12, 6))
sns.countplot(x='country', data=df, order=df['country'].value_counts().head(10).index)
plt.xticks(rotation=90)
plt.title('Distribution of Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the trend of movies and shows over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='month', hue='type', data=df)
plt.title('Trend of Movies and Shows Over Time')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

```python
# Plot the average duration by category (Top 10)
plt.figure(figsize=(12, 6))
df.groupby("categories")["duration_min"].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Average Duration by Category (Top 10)')
plt.xlabel('Category')
plt.ylabel('Average Duration (minutes)')
plt.xticks(rotation=90)
plt.show()
```
